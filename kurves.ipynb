```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
# np.random.seed(0xdeadbeef)
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import HoverTool
from scipy.interpolate import make_interp_spline
from scipy.interpolate import interp1d
output_notebook()
```

```python
x = np.array([  1,  3, 5, 7, 8, 10 ])
y = np.array([ 30, 45, 50, 68, 82, 60])
x2 = np.array([  1,  3, 5, 8, 9, 10 ])
y2 = np.array([ 25, 42, 48, 55, 63, 54])

tools = "hover, box_zoom, undo, crosshair"
plot = figure(plot_height=300)
plot.line(x, y, color='purple', line_width=2)
plot.circle(x, y, color='purple', fill_color='white', size=10)
plot.add_tools(HoverTool(tooltips=[('x','$data_x'),('y','$data_y')],mode='vline',line_policy='interp'))
#output_file('line.html')
#show(plot)

plot1 = figure()
plot1.line(x, y, color='purple', line_width=2)
plot1.circle(x, y, color='purple', fill_color='white', size=7)
plot1.line(x2, y2, color='purple', line_width=2)
plot1.circle(x2, y2, color='purple', fill_color='white', size=7)
plot1.add_tools(HoverTool(tooltips=[('x2','$data_x'),('y2','$data_y')],mode='hline',line_policy='interp'))
#output_file('line.html')

#Then, we want to find the average x and average y value for each of our estimate arrays. Just use np.mean:
midx = [np.mean([x[i], x2[i]]) for i in range(6)]
midy = [np.mean([y[i], y2[i]]) for i in range(6)]
plot1.line(midx, midy, color='red', line_width=2)
#midx and midy now represent the midpoint between our 2 estimate arrays.


X_Y_Spline = make_interp_spline(x, y)
X_ = np.linspace(x.min(), x.max(), 500)
Y_ = X_Y_Spline(X_)
plot1.line(X_, Y_)

show(plot1)

#np.interp(2, x,y)
```

```python
from bokeh.io import output_file, show
from bokeh.layouts import layout
from bokeh.models import BoxAnnotation, Toggle
from bokeh.plotting import figure

#output_file("styling_visible_annotation_with_interaction.html")

p = figure(width=1000)
blue_line = p.line(x, y, color='purple', line_width=2)
pink_line = p.line(x2, y2, color='purple', line_width=2)
points = p.circle(x, y, color='purple', fill_color='white', size=7)
points2 = p.circle(x2, y2, color='purple', fill_color='white', size=7)
midx = [np.mean([x[i], x2[i]]) for i in range(6)]
midy = [np.mean([y[i], y2[i]]) for i in range(6)]
mid_line = p.line(midx, midy, color='red', line_width=2)
ht = HoverTool(renderers=[blue_line],tooltips=[('x','$data_x'),('y','$data_y')],mode='hline',line_policy='interp')
ht2 = HoverTool(renderers=[pink_line],tooltips=[('x','$data_x'),('y','$data_y')],mode='hline',line_policy='interp')
ht_mid = HoverTool(renderers=[mid_line],tooltips=[('x','$data_x'),('y','$data_y')],mode='hline',line_policy='interp')
p.add_tools(ht,ht2,ht_mid)
           
X_Y_Spline = make_interp_spline(x, y)
# Returns evenly spaced numbers
# over a specified interval.
X_ = np.linspace(x.min(), x.max(), 500)
Y_ = X_Y_Spline(X_)
curve=p.line(X_, Y_)

# Use js_link to connect button active property to glyph visible property

toggle1 = Toggle(label="Green Box", button_type="success", active=True)
toggle1.js_link('active', blue_line, 'visible')
toggle1.js_link('active', curve, 'visible')
toggle1.js_link('active', points, 'visible')

toggle2 = Toggle(label="Pink Line", button_type="success", active=True)
toggle2.js_link('active', pink_line, 'visible')
toggle2.js_link('active', points2, 'visible')

toggle3 = Toggle(label="New Curve", button_type="success", active=True)
toggle3.js_link('active', mid_line, 'visible')

show(layout([p], [toggle1, toggle2, toggle3]))
```
